```python

import base64

# 文本加密
def encrypt_text(text):
    # 将文本转换为字节类型
    text_bytes = text.encode('utf-8')
    # 进行Base64编码
    encoded_bytes = base64.b64encode(text_bytes)
    # 将编码后的字节转换为字符串并返回
    encoded_text = encoded_bytes.decode('utf-8')
    return encoded_text

# 文本解密
def decrypt_text(encoded_text):
    # 将编码后的字符串转换为字节类型
    encoded_bytes = encoded_text.encode('utf-8')
    # 进行Base64解码
    decoded_bytes = base64.b64decode(encoded_bytes)
    # 将解码后的字节转换为字符串并返回
    decoded_text = decoded_bytes.decode('utf-8')
    return decoded_text

# 测试
text = "Hello, World!"
encoded_text = encrypt_text(text)
decoded_text = decrypt_text("aHR0cHM6Ly9naXRodWIuY29tL2NhbWVuZHVydS9zdGFibGUtZGlmZnVzaW9uLXdlYnVp")
decoded_text

from Crypto.Cipher import AES
import base64

# 使用固定的密钥，注意密钥长度必须为16字节（128位）
fixed_key = b'MyFixedEncryptio'

def pad_data(data):
    # 计算需要填充的字节数
    padding_size = AES.block_size - len(data) % AES.block_size
    # 使用ASCII码为填充值，补足数据长度
    padded_data = data + chr(padding_size) * padding_size
    return padded_data

def unpad_data(padded_data):
    # 获取填充值
    padding_size = ord(padded_data[-1])
    # 去掉填充的部分，得到原始数据
    return padded_data[:-padding_size]

def encrypt_aes_ecb(data):
    key = fixed_key
    cipher = AES.new(key, AES.MODE_ECB)
    # 填充数据
    padded_data = pad_data(data)
    # 加密
    ciphertext = cipher.encrypt(padded_data.encode('utf-8'))
    return base64.b64encode(ciphertext).decode('utf-8')

def decrypt_aes_ecb(encrypted_data):
    encrypted_data = base64.b64decode(encrypted_data)
    cipher = AES.new(fixed_key, AES.MODE_ECB)
    # 解密
    decrypted_data = cipher.decrypt(encrypted_data)
    # 去除填充
    data = unpad_data(decrypted_data.decode('utf-8'))
    return data

# 要加密的数据
data_to_encrypt = "https://github.com/deforum-art/deforum-for-automatic1111-webui"

# 进行加密
encrypted_data = encrypt_aes_ecb(data_to_encrypt)
print("Encrypted Data (ECB mode):", encrypted_data)

# 进行解密
decrypted_data_ecb = decrypt_aes_ecb(encrypted_data)
print("Decrypted Data (ECB mode):", decrypted_data_ecb)
```
